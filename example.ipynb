# BrainMatrix Example

```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
sns.set_context('notebook', font_scale=1.3)
```

## Initialzing the matrix

We begin by initializing the `BrainMatrix` object. If the neurosynth database has not
already been downloaded, it will be downloaded automatically. This took 5 minutes
with a 13 Mbps connection. Loading the dataset into memory takes around 15 seconds.
We are using a fairly large `downsample` value here to speed up the computation. You would
likely want to use sharper images (a lower `downsample`) in your research.

```python
from brain_matrix import BrainMatrix
matrix = BrainMatrix(metric='emd', downsample=10, name='example', auto_save=False)
```

Now we'll compute a full distance matrix for 6 features, a total of 15 distances.

```python
matrix.compute_distances(['syntactic', 'speech', 'semantic', 'music', 'motor', 'hand'])
```

## Visualizing the matrix

We'll visualize the resulting matrix in several ways. First, a heatmap is
just the matrix with distances encoded by color.

```python
sns.heatmap(matrix.to_dataframe())
```

A Multidimensional scaling plot shows the best embedding of the
features in a two dimensional space. This allows us to see how the
different features cluster in a reduced dimension fMRI space. Features
near each other have relatively similiar associated fMRI images.

```python
_ = matrix.plot_mds(clusters=2, dim=2)
```

Finally, a dendrogram displays a hierarchical taxonomy. It is another
way to see how the different features cluster together in
fMRI space.

```python
_ = matrix.plot_dendrogram()
```
